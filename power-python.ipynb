```python
from itertools import permutations, combinations, combinations_with_replacement


#name,limit = input().split()
name = "HACK"
limit = 2


per = [''.join(i) for i in permutations(name,int(limit))]
print('\n'.join(per))
print()

for i in range(1,int(limit)+1):
    data = [''.join(sorted(i)) for i in combinations(name,i)]
    data.sort()
    print('\n'.join(data))

print()


replace = [''.join(sorted(i)) for i in combinations_with_replacement(name,int(limit))]    
replace.sort()
print('\n'.join(replace))
```

```python
def count_substring(string, sub_string):
  total = 0
  for i in range(len(string)):
      if string[i:].startswith(sub_string):
          total += 1
  return total


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
# osamaosamaosamaosamaosama
# ama
# ==> 5
```

```python
#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
def print_formatted(number):
    w = number.bit_length()
    for num in range(1, number + 1):
        print(f"{num:>{w}d} {num:>{w}o} {num:>{w}X} {num:>{w}b}")

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
# The Minion Game
def minion_game(string):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    a = 0
    b = 0


    for i,c in enumerate(string):
        if c in vowels:
            b += len(string) - i
        else:
            a += len(string) - i
            
    if a == b:
        print('Draw')
    elif a >b:
        print(f'Stuart {a}')
    else:
        print(f'Kevin {b}')
if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
from collections import OrderedDict

def merge_the_tools(string, k):
    for i in range(0,len(string),k):
        x = string[i:i+k]
        d = OrderedDict()
        for i in x:
            if i not in d:
                d[i]=0 
            d[i]+=1 
        print(''.join(d.keys()))
    
if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
# AABCAAADA
# 3
```

```python
def reverse_words_order_and_swap_cases(sentence):
    x = sentence.split()[::-1]
    lis = [] 
    for i in x:
        lis.append(i.swapcase())
    print(' '.join(lis))

if __name__ == '__main__':

    sentence = input()

    result = reverse_words_order_and_swap_cases(sentence)

# aWESOME is cODING
# ==>> Coding IS Awesome
```

```python
class Car:
    def __init__(self, max_speed, speed_unit):
        self.max_speed = max_speed
        self.speed_unit = speed_unit
    def __repr__(self):
        return (f'Car with the maximum of {self.max_speed} {self.speed_unit}')
class Boat:
    def __init__(self, max_speed):
        self.max_speed = max_speed
    def __repr__(self):
        return (f'Boat with the maximum of {self.max_speed}')

if __name__ == '__main__':
    q = int(input())
    queries = []
    for _ in range(q):
        args = input().split()
        vehicle_type, params = args[0], args[1:]
        if vehicle_type == "car":
            max_speed, speed_unit = int(params[0]), params[1]
            vehicle = Car(max_speed, speed_unit)
        elif vehicle_type == "boat":
            max_speed = int(params[0])
            vehicle = Boat(max_speed)
            
        else:
            raise ValueError("invalid vehicle type")

        queries.append(vehicle)
    for i in queries:
        print(i)
    
# 2
# car 123 km/h
# boat 78 
# ==>> Car with the maximum of 123 km/h
# ==>> Boat with the maximum of 78
```
