```python
from itertools import permutations, combinations, combinations_with_replacement


#name,limit = input().split()
name = "HACK"
limit = 2


per = [''.join(i) for i in permutations(name,int(limit))]
print('\n'.join(per))
print()

for i in range(1,int(limit)+1):
    data = [''.join(sorted(i)) for i in combinations(name,i)]
    data.sort()
    print('\n'.join(data))

print()


replace = [''.join(sorted(i)) for i in combinations_with_replacement(name,int(limit))]    
replace.sort()
print('\n'.join(replace))
```

```python
def mutate_string(string, position, character):
    cha = list(string)
    
    cha[position] = character 
    final =''.join(cha)
    return final
if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
# osama 
# 0 U
# ==> Usama
```

```python
def count_substring(string, sub_string):
  total = 0
  for i in range(len(string)):
      if string[i:].startswith(sub_string):
          total += 1
  return total


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
# osamaosamaosamaosamaosama
# ama
# ==> 5
```

```python
#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
import textwrap

def wrap(string, max_width):
    return textwrap.wrap(string, max_width)

def wrap_fill(string, max_width):
    return textwrap.fill(string, max_width)
    

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    result_fill = wrap_fill(string, max_width)
    print(result)
    print('\n',result_fill)
```

```python
def print_formatted(number):
    w = number.bit_length()
    for num in range(1, number + 1):
        print(f"{num:>{w}d} {num:>{w}o} {num:>{w}X} {num:>{w}b}")

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```
